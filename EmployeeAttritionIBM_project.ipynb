#Train-Validation-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(input_df,target_df,stratify=target_df,shuffle=True,
                            random_state=42,test_size=0.25)
X_train,X_val,y_train,y_val=train_test_split(X_train,y_train,shuffle=True,stratify=y_train,random_state=42,test_size=0.2)
```

```python
train_df=pd.concat([X_train,y_train],axis=1)
val_df=pd.concat([X_val,y_val],axis=1)
test_df=pd.concat([X_test,y_test],axis=1)
```

##Scaling Numerical Features

```python
with pd.option_context('display.max_columns', None):
    display(raw_df[numerical_cols].describe())
```

We can see that some numerical features have values ranging up to several thousand, while others are much smaller in scale. To scale them all to the range [0,1] `MinMaxScaler` will be applied.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
scaler.fit(X_train[numerical_cols])
```

```python
X_train[numerical_cols]=scaler.transform(X_train[numerical_cols])
X_val[numerical_cols]=scaler.transform(X_val[numerical_cols])
X_test[numerical_cols]=scaler.transform(X_test[numerical_cols])
```

```python
train_df[numerical_cols].describe()
```

We have prevented data leakage by fitting the `MinMaxScaler` with only training data. This will make sure that validation and test data remain truly unseen.

##Encoding the Cateogrical Columns

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
train_df[categorical_cols].nunique()
```

```python
raw_df[categorical_cols].nunique()
```

```python
encoder = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
encoder.fit(train_df[categorical_cols])
```

```python
encoded_cols=encoder.get_feature_names_out(categorical_cols)
encoded_cols
```

```python
encoded_df=pd.DataFrame(
    encoder.transform(train_df[categorical_cols]),
    index=train_df.index,
    columns=encoded_cols
)
train_df=pd.concat([train_df.drop(columns=categorical_cols),encoded_df],axis=1)
```

```python
encoded_df=pd.DataFrame(
    encoder.transform(val_df[categorical_cols]),
    index=val_df.index,
    columns=encoded_cols
)
val_df=pd.concat([val_df.drop(columns=categorical_cols),encoded_df],axis=1)

encoded_df=pd.DataFrame(
    encoder.transform(test_df[categorical_cols]),
    index=test_df.index,
    columns=encoded_cols
)
test_df=pd.concat([test_df.drop(columns=categorical_cols),encoded_df],axis=1)
```

```python
train_df
```

```python
X_train = train_df[numerical_cols+list(encoded_cols)]
X_val=val_df[numerical_cols+list(encoded_cols)]
X_test=test_df[numerical_cols+list(encoded_cols)]
```

```python
y_train=train_df['Attrition']
y_val=val_df['Attrition']
y_test=test_df['Attrition']
```

#Model Training

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(model.predict(X_train),y_train))
```

```python
print(classification_report(y_val,model.predict(X_val)))
```

86% is a good accuracy, but the model is only slightly better than 'dumb' model that always predicts 'NO'.

```python
model.predict_proba(X_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_train,model.predict(X_train),normalize='true')
sns.heatmap(cm,cmap='Reds',annot=True)
plt.xlabel('Predicted')
plt.ylabel('True')
```

Although the model achieves an accuracy of 86%, the confusion matrix reveals a high false-negative rate for the minority class, `Attrition=Yes`. This means our logistic regression model is good at predicting who doesn't attrit. But business-wise, HR cares about employes that might quit. So, even though this model isn't terrible ML wise, this is not enough for the real world problem we aim to solve.

The significant amount of false-negatives are probably being caused by the class imbalance. Considering this, let's test a new model were classes are balanced.

```python
model = LogisticRegression(class_weight='balanced')
model.fit(X_train,y_train)
```

```python
print(classification_report(y_train,model.predict(X_train)))
```

```python
print(classification_report(y_val,model.predict(X_val)))
```

```python
cm=confusion_matrix(y_val,model.predict(X_val),normalize='true')
sns.heatmap(cm,cmap='Reds',annot=True)
plt.xlabel('Predicted')
plt.ylabel('True')
```

After class-balancing, we have managed to achieve a significant increase in recall,i.e a lot fewer false negatives. But the accuracy of the model has fallen below that of a 'dumb' model. This suggests that a linear model may not be the best approach to solving this problem. So let's approach this problem using a Gaussian Classifier, which can identify nonlinear patterns.

### Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
print(classification_report(y_train,model.predict(X_train)))
```

```python
print(classification_report(y_val,model.predict(X_val)))
```

The `RandomForestClassifier` seems to be overfitting on the training data. This can be prevented by limting parameters such as maximum leaves and maximum depth.

```python
model = RandomForestClassifier(n_jobs=-1,max_depth=5)
model.fit(X_train,y_train)
```

```python
print(classification_report(y_train,model.predict(X_train)))
```

```python
print(classification_report(y_val,model.predict(X_val)))
```

```python
model = RandomForestClassifier(n_jobs=-1,max_depth=15,max_leaf_nodes=50)
model.fit(X_train,y_train)
```

```python
print(classification_report(y_val,model.predict(X_val)))
```

Let us try to tune the hyper-parameters `max_depth` and `max_leaf_nodes` by maximising validation recall.

```python
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
```

```python
max_recall=0
best_model=None
for x in range(1,30):
  for y in range(2,200):
    model=RandomForestClassifier(n_jobs=-1,max_depth=x,max_leaf_nodes=y,random_state=42)
    model.fit(X_train,y_train)
    y_pred=model.predict(X_val)
    cm = confusion_matrix(y_val,y_pred)
    recall = recall_score(y_val, y_pred, zero_division=0)
    if(recall>max_recall):
      max_recall=recall
      best_model=model
```

```python
print(recall)
```

Even after trying to tune the hyperparameters `max_depth` and `max_leaf_nodes`, the maximum recall is only around **0.2**. So let's try altering the decision threshold to see if it improves recall.

```python
y_proba = best_model.predict_proba(X_val)[:, 1]

threshold = 0.3
y_pred = (y_proba >= threshold).astype(int)
```

```python
print(recall_score(y_val,y_pred,zero_division=0))
print(accuracy_score(y_val,y_pred))
```

Here, we can clearly observer that decreasing the threshold significantly improved the recall score which is important for the real world problem we aim to solve. But the accuracy decreased too, so lets find the best threshold where we can get a good recall score while maintaining a decent accuracy too.

```python
t=0
max_recall=0
for x in range(10,31):
  threshold=x/100
  y_proba = best_model.predict_proba(X_val)[:, 1]
  y_pred = (y_proba >= threshold).astype(int)
  recall=recall_score(y_val,y_pred,zero_division=0)
  if(recall>max_recall and accuracy_score(y_val,y_pred)>0.7):
      max_recall=recall
      t=x
print(f"Optimal Recall:{max_recall}")
print(f"threshold:{t/100}")
```

It has been found that the best possible recall is approximately 0.72 while maintaing an optimal accuracy above 70%.

Lets try experimenting with a few more hyperparameters to maximise recall. Let's begin by setting custom class-weights as balanced class weights might harm the model.

```python
model = RandomForestClassifier(n_jobs=-1,max_depth=15,max_leaf_nodes=50,n_estimators=500,class_weight={0:1,1:2},random_state=42)
model.fit(X_train,y_train)
```

```python
print(classification_report(y_train,model.predict(X_train)))
```

```python
print(classification_report(y_val,model.predict(X_val)))
```

```python
y_proba = model.predict_proba(X_val)[:,1]
threshold=0.3
y_pred=(y_proba>threshold).astype(int)
print(recall_score(y_val,y_pred))
print(accuracy_score(y_val,y_pred))
```

Balancing the class weights and increasing the no.of estimators in the Random Forest significantly improved the recall score while maintaing a better accuracy. Now, let's find the best threshold while maintaining a optimal accuracy.

```python
t=0
max_recall=0
for x in range(10,31):
  threshold=x/100
  y_proba = model.predict_proba(X_val)[:, 1]
  y_pred = (y_proba >= threshold).astype(int)
  recall=recall_score(y_val,y_pred,zero_division=0)
  if(recall>max_recall and accuracy_score(y_val,y_pred)>0.65):
      max_recall=recall
      t=x
print(f"Optimal Recall:{max_recall}")
print(f"threshold:{t/100}")
```
